```python
#before 

# Style
sns.set(style="whitegrid")

# Timestamp Conversion
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df['Year'] = df['Timestamp'].dt.year

# CREATE SUBPLOTS 
plt.figure(figsize=(18, 15))

# 1. Rating Distribution
plt.subplot(3,3,1)
sns.histplot(df['Rating'], kde=True)
plt.title("Rating Distribution")

# 2. Genre
plt.subplot(3,3,2)
df['Genre'].value_counts().head(10).plot(kind='bar')
plt.title("Top Genres")

# 3. Country
plt.subplot(3,3,3)
df['Country'].value_counts().head(10).plot(kind='bar')
plt.title("Top Countries")

# 4. Release Year
plt.subplot(3,3,4)
df['Release_Year'].value_counts().sort_index().plot()
plt.title("Movies Over Years")

# 5. Category
plt.subplot(3,3,5)
sns.countplot(x=df['Category'])
plt.title("Category Distribution")


# 6. Language
plt.subplot(3,3,7)
df['Language'].value_counts().head(10).plot(kind='bar')
plt.title("Top Languages")

# 7. Dubbed Languages
plt.subplot(3,3,8)
df['DubbedLanguages'].value_counts().head(10).plot(kind='bar')
plt.title("Dubbed Languages")

# 8. User Activity Over Time
plt.subplot(3,3,9)
df['Year'].value_counts().sort_index().plot()
plt.title("User Activity Over Time")

plt.tight_layout()
plt.show()
```

```python
# FIX DURATION COLUMN
df['Duration'] = df['Duration'].astype(str).str.extract('(\d+)')
df['Duration'] = pd.to_numeric(df['Duration'], errors='coerce')
df = df.dropna(subset=['Duration'])
data = df['Duration']

# TRANSFORMATIONS

log_data = np.log1p(data)
exp_data = np.exp(data / data.max())


# PLOTS
# =============================
plt.figure(figsize=(18,10))

# Original
plt.subplot(2,2,1)
sns.histplot(data, kde=True)
plt.title("Original Distribution")

# Log
plt.subplot(2,2,2)
sns.histplot(log_data, kde=True)
plt.title("Log Transformed")

# Exponential
plt.subplot(2,2,3)
sns.histplot(exp_data, kde=True)
plt.title("Exponential Transform")

# Bell Curve
plt.subplot(2,2,4)
x = np.linspace(-3, 3, 100)
plt.plot(x, norm.pdf(x))
plt.title("Ideal")
```

```python
#handling null value
for x in ["Category","Timestamp","Country","Release_Year","Genre","Language","DubbedLanguages","Duration","Year"]:
    df[x].fillna(df[x].mode()[0],inplace=True)
df["Rating"].fillna(df["Rating"].mean(),inplace=True)
```

```python
sequence = pd.Series('s' + (df.index + 1).astype(str), index=df.index)
df['Movie_ID'] = df['Movie_ID'].fillna(sequence)
df.isna().sum()
```

```python
sequence = pd.Series('P' + (df.index + 1).astype(str), index=df.index)
df['User_ID'] = df['User_ID'].fillna(sequence)
df.isna().sum()
```

```python
null_titles = df[df['Title'].isnull()]
print(f"Total missing titles: {len(null_titles)}")
print("\nsome rows where Title is missing:")
print(null_titles[['Movie_ID', 'Category', 'Title']].head())
unique_titles = pd.Series('Untitled_' + (df.index + 1).astype(str), index=df.index)
df['Title'] = df['Title'].fillna(unique_titles)
```

```python
numeric_cols = df.select_dtypes(include=np.number)
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_iqr = ((numeric_cols < lower_bound) | 
                (numeric_cols > upper_bound))
df= df[~outliers_iqr.any(axis=1)]
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = ['User_ID','Movie_ID','Category','Title','Country','Genre','Language','DubbedLanguages']
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
from sklearn.preprocessing import StandardScaler
X = df.drop(["Timestamp", "Duration"], axis=1)
# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df["preference"] = df["Rating"].apply(lambda x: 1 if x > 7 else 0)
```

```python
y=df["preference"]
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
from sklearn.linear_model import LogisticRegression

# Model define
model = LogisticRegression(max_iter=1000)

# SFS apply
sfs = SequentialFeatureSelector(model, n_features_to_select=5,direction='forward')
sfs.fit(X_scaled, y)
selected_features_forward = X.columns[sfs.get_support()]
selected_features_forward=list(selected_features_forward)
print("Selected Features:",selected_features_forward)
```

```python
sfs2 = SequentialFeatureSelector(model, n_features_to_select=5,direction='backward')
sfs2.fit(X_scaled, y)
selected_features_backward = X.columns[sfs2.get_support()]

print("Selected Features:", list(selected_features_backward))
```

```python
common_features_lr=(set(selected_features_forward )| set(selected_features_backward))
common_features_lr
```

```python
from sklearn.neighbors import KNeighborsClassifier
model =KNeighborsClassifier(n_neighbors=6)
sfs2 = SequentialFeatureSelector(model, 
                                n_features_to_select="auto",  
                                direction='forward')
sfs2.fit(X_scaled, y)
selected_features_forward = X.columns[sfs2.get_support()]
print("Selected Features:", list(selected_features_forward))
```

```python
sfs2 = SequentialFeatureSelector(model,  n_features_to_select="auto",direction='backward')
sfs2.fit(X_scaled, y)
selected_features_backward = X.columns[sfs2.get_support()]
print("Selected Features:", list(selected_features_backward))
```

```python
common_features_knn=set(set(selected_features_forward) | set(selected_features_backward))
common_features_knn
```
