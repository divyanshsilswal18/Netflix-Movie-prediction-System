```python
#before 

# Style
sns.set(style="whitegrid")

# Timestamp Conversion
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df['Year'] = df['Timestamp'].dt.year

# CREATE SUBPLOTS 
plt.figure(figsize=(18, 15))

# 1. Rating Distribution
plt.subplot(3,3,1)
sns.histplot(df['Rating'], kde=True)
plt.title("Rating Distribution")

# 2. Genre
plt.subplot(3,3,2)
df['Genre'].value_counts().head(10).plot(kind='bar')
plt.title("Top Genres")

# 3. Country
plt.subplot(3,3,3)
df['Country'].value_counts().head(10).plot(kind='bar')
plt.title("Top Countries")

# 4. Release Year
plt.subplot(3,3,4)
df['Release_Year'].value_counts().sort_index().plot()
plt.title("Movies Over Years")

# 5. Category
plt.subplot(3,3,5)
sns.countplot(x=df['Category'])
plt.title("Category Distribution")


# 6. Language
plt.subplot(3,3,7)
df['Language'].value_counts().head(10).plot(kind='bar')
plt.title("Top Languages")

# 7. Dubbed Languages
plt.subplot(3,3,8)
df['DubbedLanguages'].value_counts().head(10).plot(kind='bar')
plt.title("Dubbed Languages")

# 8. User Activity Over Time
plt.subplot(3,3,9)
df['Year'].value_counts().sort_index().plot()
plt.title("User Activity Over Time")

plt.tight_layout()
plt.show()
```

```python
# FIX DURATION COLUMN
df['Duration'] = df['Duration'].astype(str).str.extract('(\d+)')
df['Duration'] = pd.to_numeric(df['Duration'], errors='coerce')
df = df.dropna(subset=['Duration'])
data = df['Duration']

# TRANSFORMATIONS

log_data = np.log1p(data)
exp_data = np.exp(data / data.max())


# PLOTS
# =============================
plt.figure(figsize=(18,10))

# Original
plt.subplot(2,2,1)
sns.histplot(data, kde=True)
plt.title("Original Distribution")

# Log
plt.subplot(2,2,2)
sns.histplot(log_data, kde=True)
plt.title("Log Transformed")

# Exponential
plt.subplot(2,2,3)
sns.histplot(exp_data, kde=True)
plt.title("Exponential Transform")

# Bell Curve
plt.subplot(2,2,4)
x = np.linspace(-3, 3, 100)
plt.plot(x, norm.pdf(x))
plt.title("Ideal")
```
